```python
import pandas as pd

df = pd.read_csv('new_logs.csv')
df.Timestamp = pd.to_datetime(df.Timestamp)
```

```python
df
```

```python
def filter_timestamps_in_range(df, start_time, end_time):
    # Convert start and end times to Timestamp objects
    start_timestamp = pd.to_datetime(start_time)
    end_timestamp = pd.to_datetime(end_time)
    
    # Filter the DataFrame to include only timestamps within the specified range
    filtered_df = df[
        (df['Timestamp'] >= start_timestamp) &
        (df['Timestamp'] <= end_timestamp)
    ]
    
    return filtered_df['Details'].unique()
```

```python
x = filter_timestamps_in_range(df, '2022-04-01', '2022-04-15')
```
